### 1. What types are involved in the following expressions? What data type will each expression evaluate to?

`8 * 2.5`

```python
# Your code here
variable1 = 8 * 2.5
print(variable1)
type(variable1)

#The answer is "float"
```

`9 / 2`

```python
# Your code here
variable2 = 9 / 2
print(variable2, type(variable2))
```

`9 // -2`

```python
# Your code here
variable3 = 9 // -2
print(variable3, type(variable3))
```

`1.5 * 2 >= 7 - 3`

```python
# Your code here
variable4 = 1.5 * 2 >= 7 - 3
print(variable4, type(variable4))
```
